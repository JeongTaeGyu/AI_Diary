```python
model_path = os.listdir('/content/drive/MyDrive/sd_model_fintuning_LoRA/fintune_model_ghibli_v2')[-1] #FineTuning Model Path
pipe = StableDiffusionImg2ImgPipeline.from_pretrained("stable-diffusion-v1-5/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe.unet.load_attn_procs('/content/drive/MyDrive/sd_model_fintuning_LoRA/fintune_model_ghibli_v2/'+model_path)
pipe.to("cuda")

#positive prompt
prompt = '''(masterpiece, best quality),
((ghibli)), (studios ghibli), ((ghibli style)), Retro Ghibli scene,
family eating pizza together at a home table, 1gril , mom and dad, upper body, focus on face
smile, highly detailed facial expressions, joyful atmosphere, warm and soft lighting, hartwarming moments,
relaxed and happy atmosphere,
  '''


#negative prompt
neg_prompt='''FastNegativeV2,(bad-artist:1.0), (loli:1.2),
(worst quality, low quality:1.4), (bad_prompt_version2:0.8),
bad-hands-5,lowres, bad anatomy, ((bad face)), ((missing face)), ((text)), (watermark),
error, missing fingers, extra digit, fewer digits, cropped,
worst quality, low quality, normal quality, ((username)), blurry,
 (extra limbs), bad-artist-anime, badhandv4, EasyNegative,
 ng_deepnegative_v1_75t, verybadimagenegative_v1.3, BadDream,
(three hands:1.1),(three legs:1.1),(more than two hands:1.4),
(more than two legs,:1.2),badhandv4,EasyNegative,ng_deepnegative_v1_75t,verybadimagenegative_v1.3,(worst quality, low quality:1.4),text,words,logo,watermark,
'''
init_image = Image.open("/content/drive/MyDrive/egimg.png").convert("RGB")
init_image = init_image.resize((768, 512))

image = pipe(prompt, negative_prompt=neg_prompt, image=init_image, num_inference_steps=50, guidance_scale=8,
             ).images[0]
#image.save("/content/drive/MyDrive/sd_model_fintuning_LoRA/image/tfimagetest1.png")
image
```
