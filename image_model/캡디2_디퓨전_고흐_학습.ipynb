```python
import os
import time
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.common.by import By
from selenium.common.exceptions import WebDriverException, NoSuchElementException, TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import requests

# 크롬 드라이버 설정 함수
def set_chrome_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # GUI 없이 실행
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--disable-blink-features=AutomationControlled')
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)
    return driver

# 이미지 다운로드 함수 정의
def download_image(url, save_path):
    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(url, headers=headers, stream=True)
        response.raise_for_status()
        with open(save_path, 'wb') as out_file:
            for chunk in response.iter_content(1024):
                out_file.write(chunk)
    except requests.exceptions.RequestException as e:
        print(f"이미지 다운로드 실패: {e}")

# 메인 크롤링 스크립트 작성
try:
    # 크롬 드라이버 설정
    driver = set_chrome_driver()

    # 이미지 저장 폴더 설정
    save_folder = '/content/drive/MyDrive/ddata/Sample/gogh_images'
    os.makedirs(save_folder, exist_ok=True)

    # 첫 작품 페이지로 이동
    first_artwork_url = "https://www.wikiart.org/en/vincent-van-gogh/field-of-grass-with-dandelions-and-tree-trunks"
    driver.get(first_artwork_url)
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))  # 페이지 로딩 대기

    # 작품 수 초기화
    downloaded_count = 0

    # 각 작품 페이지로 이동하여 큰 이미지 다운로드
    while True:
        try:
            artwork_html = driver.page_source
            artwork_soup = BeautifulSoup(artwork_html, 'html.parser')

            # 큰 이미지 다운로드
            main_image = artwork_soup.find("img", {'itemprop': 'image'})
            if main_image and 'src' in main_image.attrs:
                image_url = main_image['src'].split("!")[0] + "!Large.jpg"
                image_name = os.path.basename(image_url.split("!")[0])
                save_path = os.path.join(save_folder, image_name)

                if not os.path.exists(save_path):
                    download_image(image_url, save_path)
                    print(f"이미지 다운로드 완료: {image_name}")
                    downloaded_count += 1

            # 다음 작품 페이지로 이동
            next_link = artwork_soup.find("a", class_="btn-overlay-container-artwork-prev")
            if next_link and 'onclick' in next_link.attrs:
                next_artwork_url = next_link['onclick'].split("nextArtworkClick(event, '")[1].split("')")[0]
                driver.get(f"https://www.wikiart.org{next_artwork_url}")
                WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))  # 페이지 로딩 대기
            else:
                print("다음 페이지로 이동 재시도")
                time.sleep(5)
                next_link = artwork_soup.find("a", class_="btn-overlay-container-artwork-prev")
                if next_link and 'onclick' in next_link.attrs:
                    next_artwork_url = next_link['onclick'].split("nextArtworkClick(event, '")[1].split("')")[0]
                    driver.get(f"https://www.wikiart.org{next_artwork_url}")
                    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))  # 페이지 로딩 대기
                else:
                    print("더이상 작품이 없습니다.")
                    break
        except (WebDriverException, NoSuchElementException, TimeoutException) as e:
            print(f"웹 드라이버 오류 발생: {e}. 페이지를 다시 로드합니다.")
            time.sleep(5)
            driver.refresh()
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
        except Exception as e:
            print(f"오류 발생: {e}. 다음 페이지로 이동을 시도합니다.")
            time.sleep(5)
            continue

except Exception as e:
    print(f"크롤링 중 오류 발생: {e}")

finally:
    # 드라이버 종료
    if 'driver' in locals():
        driver.quit()
    print("모든 이미지 다운로드 완료.")
    print(f"다운로드 완료된 작품 수: {downloaded_count}")
```

```python
from transformers import AutoProcessor, Blip2ForConditionalGeneration,BlipForQuestionAnswering,Blip2Processor,BlipForConditionalGeneration
import torch

#processor = AutoProcessor.from_pretrained("Salesforce/blip2-opt-2.7b")
model = Blip2ForConditionalGeneration.from_pretrained("Salesforce/blip2-opt-2.7b", torch_dtype=torch.float16)
processor = Blip2Processor.from_pretrained("Salesforce/blip2-opt-2.7b")
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
import torch
import gc

torch.cuda.empty_cache()
gc.collect()
```

```python
import os
import pandas as pd

prompt = "a painting of"
caption=[]
img_name=[]

file_list=os.listdir('/content/drive/MyDrive/ddata/Sample/gogh_images')

for i in range(len(file_list)):

  #이미지 이름 저장
  img_name.append(file_list[i])

  #이미지 업로드
  img=Image.open('/content/drive/MyDrive/ddata/Sample/gogh_images/'+file_list[i]).convert('RGB')
  img=img.resize((256,256))

  gc.collect()
  torch.cuda.empty_cache()

  inputs = processor(img, text=prompt, return_tensors="pt").to(device, torch.float16)

  generated_ids = model.generate(**inputs, max_new_tokens=600)
  generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()
  caption.append(generated_text)

  gc.collect()
  torch.cuda.empty_cache()

#이미지 이름과 caption저장
image_df_gogh=pd.DataFrame({'image':img_name, 'text':caption})
image_df_gogh
```
